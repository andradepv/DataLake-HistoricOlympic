```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import json

bronze_path = '../../bronze/' 
gold_medalhas_path = './'          

df_paris = pd.read_parquet(f"{bronze_path}olympic_paris_medals.parquet")

quadro = df_paris.groupby('country')['medal_type'].value_counts().unstack(fill_value=0)

if 'Gold Medal' in quadro.columns:
    quadro = quadro.rename(columns={'Gold Medal': 'Ouro', 'Silver Medal': 'Prata', 'Bronze Medal': 'Bronze'})
quadro = quadro.sort_values(by=['Ouro', 'Prata', 'Bronze'], ascending=False).reset_index()

top_10 = quadro.head(10)

top_10.to_parquet(f"{gold_medalhas_path}quadro_medalhas.parquet", index=False)
top_10.to_csv(f"{gold_medalhas_path}medalhas_summary.csv", index=False)

plt.figure(figsize=(10, 6))
plt.bar(top_10['country'], top_10['Ouro'], color='gold', label='Ouro')
plt.bar(top_10['country'], top_10['Prata'], color='silver', bottom=top_10['Ouro'], label='Prata')
plt.bar(top_10['country'], top_10['Bronze'], color='#cd7f32', bottom=top_10['Ouro'] + top_10['Prata'], label='Bronze')

plt.title('Top 10 Países - Quadro de Medalhas (Paris 2024)')
plt.xlabel('País')
plt.ylabel('Total de Medalhas')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig(f"{gold_medalhas_path}medalhas_plot.png")
plt.close()

metadados = {
  "nome_dataset": "quadro_medalhas_top10",
  "fonte": "Camada Bronze (olympic_paris_medals)",
  "descricao": "Quadro de medalhas agregado e ordenado para os 10 melhores países de Paris 2024.",
  "campos_principais": ["country", "Ouro", "Prata", "Bronze"],
  "data_coleta": "2026-03-30"
}
with open(f"{gold_medalhas_path}metadata.json", 'w', encoding='utf-8') as f:
    json.dump(metadados, f, indent=4, ensure_ascii=False)

print("✅ Análise de Medalhas corrigida! Valores de Paris 2024 estão exatos agora.")
```
