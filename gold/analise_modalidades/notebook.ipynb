```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import json

bronze_path = '../../bronze/' 
gold_modalidades_path = './'       

df_paris = pd.read_parquet(f"{bronze_path}olympic_paris_medals.parquet")

modalidades = df_paris['discipline'].value_counts().reset_index()
modalidades.columns = ['Esporte', 'Total_Medalhas']

top_10_esportes = modalidades.head(10)

top_10_esportes.to_parquet(f"{gold_modalidades_path}modalidades_summary.parquet", index=False)
top_10_esportes.to_csv(f"{gold_modalidades_path}modalidades_summary.csv", index=False)

plt.figure(figsize=(10, 6))

plt.barh(top_10_esportes['Esporte'], top_10_esportes['Total_Medalhas'], color='#4682B4')

plt.title('Top 10 Modalidades que Mais Distribuíram Medalhas (Paris 2024)')
plt.xlabel('Total de Medalhas')
plt.ylabel('Modalidade (Esporte)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(f"{gold_modalidades_path}modalidades_plot.png")
plt.close()


metadados = {
  "nome_dataset": "modalidades_summary_top10",
  "fonte": "Camada Bronze (olympic_paris_medals)",
  "descricao": "Contagem total de medalhas distribuídas agrupadas por modalidade (esporte) em Paris 2024.",
  "campos_principais": ["Esporte", "Total_Medalhas"],
  "data_coleta": "2026-03-30"
}

with open(f"{gold_modalidades_path}metadata.json", 'w', encoding='utf-8') as f:
    json.dump(metadados, f, indent=4, ensure_ascii=False)

print("✅ Análise de Modalidades concluída! Dataset, Gráfico e Metadados gerados com sucesso.")
```
